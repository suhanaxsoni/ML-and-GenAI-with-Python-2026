```python
# Find sum of first 10 natural numbers.
sum=0
for i in range(1,11):
  sum=sum+i
print(f"sum of first ten natural numbers: {sum}")
```

```python
# Find factorial of a number.
n=int(input("enter a number: "))
fact=1
for i in range(1,n+1):
  fact=fact*i
print(f"factorial of n is: {fact}")
```

```python
#Print Fibonacci Series
n=int(input("enter the number uptil which you want the fibonacci series: "))
a=0
b=1
print(a)
i=2
while(i!=n):
  print(b,)
  a,b=b,a+b
  i=i+1
print(b)
```
