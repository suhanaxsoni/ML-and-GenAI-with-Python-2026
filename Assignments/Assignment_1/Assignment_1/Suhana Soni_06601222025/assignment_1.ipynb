```python
#find the area of rectangle
length=int(input("enter the length: "))
breadth=int(input("enter the breadth: "))
area=length*breadth
print("area of the rectangle is: ",area)
```

```python
# find simple interest
p=int(input("enter the principle amount: "))
r=int(input("enter the rate of interest: "))
t=int(input("enter the time period: "))
si=(p*r*t)/100
print("simple interest is: ",si)
```

```python
# celsius to fahrenheit
celsius=float(input("enter the temperature in celsius: "))
fahrenheit=(celsius*9/5)+32
print("temperature in fahrenheit is: ",fahrenheit)
```

```python
# average of three numbers
a=int(input("enter the first number: "))
b=int(input("enter the second number: "))
c=int(input("enter the third number: "))
average=(a+b+c)/3
print("average of three numbers is: ",average)
```

```python
# find square and cube of a number
number=int(input("enter the number: "))
square=number**2
cube=pow(number,3)
print("square of the number is: ",square)
print("cube of the number is: ",cube)
```

```python
# swap two numbers without third variable
a=int(input("enter the first number: "))
b=int(input("enter the second number: "))
a=a+b
b=a-b
a=a-b
print(f"swapped numbers are: {a},{b}")
```

```python
# create a student report that takes student details using input stores marks and calculates total and percentage
name=input("enter the name of the student: ")
roll_no=int(input("enter the roll number of the student: "))
math_marks=int(input("enter the marks obtained in maths: "))
science_marks=int(input("enter the marks obtained in science: "))
english_marks=int(input("enter the marks obtained in english: "))
total_marks=math_marks+science_marks+english_marks
percentage=(total_marks/300)*100
print(f"student name: {name}")
print(f"roll number: {roll_no}")
print(f"total marks out of 300: {total_marks}")
print(f"percentage: {percentage}")
```
