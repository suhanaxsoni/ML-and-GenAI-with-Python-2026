```python
# function to print 10 natural numbers
def natural_numbers():
  for i in range(1,11):
    print(i)
  return
natural_numbers()
```

```python
# function to calculate sum of first n natural numbers
n=int(input("Enter the value of n: "))
def sum(n):
  sum=0
  for i in range(1,n+1):
    sum=sum+i
  return sum
sum(n)
```

```python
# function to reverse a number
n=int(input("Enter the number: "))
def reverse(n):
  rev=0
  while(n>0):
    dig=n%10
    rev=rev*10+dig
    n=n//10
  return rev
reverse(n)
```

```python
# function to count digits in a number
n=int(input("enter the number: "))
def count(n):
  count=0
  while(n>0):
    count=count+1
    n=n//10
  return count
count(n)
```

```python
# function to check palindrome number
n=int(input("Enter the number: "))
def check_palindrome(n):
  rev=0
  temp=n
  while(n>0):
    digit=n%10
    rev=rev*10+digit
    n=n//10
  if(temp==rev):
    return True
  else:
    return False
check_palindrome(n)
```

```python
# function for fibonacci series
n=int(input("Enter the value of n: "))
def fib(n):
  a=0
  b=1
  while(b<=n):
    print(a)
    b,a=a+b,b
  return
fib(n)
```

```python
# calculator
operation=int(input("Enter 1 for addition, 2 for subtraction, 3 for multiplication, 4 for division: "))
a=int(input("Enter the first number: "))
b=int(input("Enter the second number:"))
def calc(operation,a,b):
  if operation==1:
    return a+b
  elif operation==2:
    return a-b
  elif operation==3:
    return a*b
  elif operation==4:
    return a/b
  else:
    return "Invalid operation"
calc(operation,a,b)
```
